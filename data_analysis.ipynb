```python
import matplotlib.pyplot as plt
import pandas as pd
% matplotlib inline
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
df = pd.read_csv('data/dataset_default_vals.csv', sep='\\t')
```

```python
df[:5]
```

```python
fig, ax = plt.subplots()
ax.scatter(df["px_gt"], df["py_gt"], alpha=0.2, label="ground truth")
ax.scatter(df["pos1_est"], df["pos2_est"], alpha=0.3, marker='x', c='r', label="estimates")
ax.legend(loc='upper left')
```

```python
fig, ax = plt.subplots()
ax.plot(df['NIS'], 'rx')
ax.axhline(7.8, color='g')
ax.set_xlabel('time')
ax.set_ylabel('chi^2')
ax.legend(loc='upper left')
```

```python
plt.plot(df['rmse_x'], label='rmse_x', color='r')
plt.plot(df['rmse_y'], label='rmse_y', color='g')
plt.plot(df['rmse_vx'], label='rmse_vx', color='b')
plt.plot(df['rmse_vy'], label='rmse_vy', color='y')
plt.ylabel('RMSE')
plt.xlabel('time')
plt.legend(loc='center right')
```
