```python
import matplotlib.pyplot as plt
import pandas as pd
% matplotlib inline

plt.rcParams['figure.figsize'] = (16, 6)
```

```python
df_default = pd.read_csv('data/dataset_default_vals.csv', sep='\\t', engine='python')
df_tuned = pd.read_csv('data/dataset_tuned_vals.csv', sep='\\t', engine='python')
```

```python
df_default_title = "std_a_ = 30 / std_yawdd_ = 30"
df_tuned_title = "std_a_ = 0.2 / std_yawdd_ = 0.2"
```

```python
df_default[:2]
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True)

df = df_default
ax1.set_title(df_default_title)
ax1.scatter(df["px_gt"], df["py_gt"], alpha=0.2, label="ground truth")
ax1.scatter(df["pos1_est"], df["pos2_est"], alpha=0.3, marker='x', c='r', label="estimates")
ax1.legend(loc='upper left')

df = df_tuned
ax2.set_title(df_tuned_title)
ax2.scatter(df["px_gt"], df["py_gt"], alpha=0.2, label="ground truth")
ax2.scatter(df["pos1_est"], df["pos2_est"], alpha=0.3, marker='x', c='r', label="estimates")
ax2.legend(loc='upper left')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True)

df = df_default
ax1.set_title(df_default_title)
ax1.plot(df['NIS'], 'rx')
ax1.axhline(7.8, color='g')
ax1.set_xlabel('time')
ax1.set_ylabel('chi^2')
ax1.legend(loc='upper right')

df = df_tuned
ax2.set_title(df_tuned_title)
ax2.plot(df['NIS'], 'rx')
ax2.axhline(7.8, color='g')
ax2.set_xlabel('time')
ax2.set_ylabel('chi^2')
ax2.legend(loc='upper right')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True)

df = df_default
ax1.set_title(df_default_title)
ax1.plot(df['rmse_x'], label='rmse_x', color='r')
ax1.plot(df['rmse_y'], label='rmse_y', color='g')
ax1.plot(df['rmse_vx'], label='rmse_vx', color='b')
ax1.plot(df['rmse_vy'], label='rmse_vy', color='y')
ax1.set_ylabel('RMSE')
ax1.set_xlabel('time')
ax1.legend(loc='upper right')

df = df_tuned
ax2.set_title(df_tuned_title)
ax2.plot(df['rmse_x'], label='rmse_x', color='r')
ax2.plot(df['rmse_y'], label='rmse_y', color='g')
ax2.plot(df['rmse_vx'], label='rmse_vx', color='b')
ax2.plot(df['rmse_vy'], label='rmse_vy', color='y')
ax2.set_ylabel('RMSE')
ax2.set_xlabel('time')
ax2.legend(loc='upper right')
```

```python
"'Default params': [{}, {}, {}, {}]".format(df_default['rmse_x'][df.index[-1]], df_default['rmse_y'][df.index[-1]], 
                              df_default['rmse_vx'][df.index[-1]], df_default['rmse_vy'][df.index[-1]])
```

```python
"'Final results': [{}, {}, {}, {}]".format(df_tuned['rmse_x'][df.index[-1]], df_tuned['rmse_y'][df.index[-1]], 
                              df_tuned['rmse_vx'][df.index[-1]], df_tuned['rmse_vy'][df.index[-1]])
```
